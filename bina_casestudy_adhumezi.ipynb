### 3.1.2 Daten aufbereiten

Folgendes kann gemacht werden:

*   Unnötige Zeilen, Werte entfernen

*   Überschriften ändern

```python
def clean_and_validate_years(df):
    """
    Reinigt und validiert die 'Jahr' Spalte im DataFrame, um sicherzustellen, dass alle Jahr-Werte
    als ganze Zahlen innerhalb des Bereichs 1980 bis 2050 liegen.
    """
    try:
        df['Jahr'] = pd.to_numeric(df['Jahr'], errors='raise').astype(int)
        if not df['Jahr'].between(1980, 2050).all():
            raise ValueError("Einige 'Jahr' Werte liegen außerhalb des Bereichs 1980 bis 2050.")
    except Exception as e:
        raise ValueError(f"Fehler bei der Validierung der 'Jahr' Spalte: {e}")

    return df
```

```python
def clean_tj_column(df):
    """
    Reinigt die 'TJ' Spalte im DataFrame, indem Werte zu numerischen Werten konvertiert werden,
    alle negativen Werte in positive umgewandelt und in ganze Zahlen umgewandelt werden.
    """
    try:
        original_data = df['TJ'].copy()
        df['TJ'] = pd.to_numeric(df['TJ'], errors='coerce').abs().dropna().astype(int)
        if df['TJ'].isnull().any():
            raise ValueError("Nicht alle Werte konnten in positive ganze Zahlen konvertiert werden.")
        if not df['TJ'].equals(original_data):
            print("Änderungen wurden an der 'TJ' Spalte vorgenommen.")
    except Exception as e:
        raise ValueError(f"Fehler bei der Bereinigung der 'TJ' Spalte: {e}")

    return df
```

```python
def check_required_columns(df, required_columns):
    """
    Überprüft, ob der gegebene DataFrame alle erforderlichen Spalten enthält.
    """
    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        raise ValueError(f"Fehlende erforderliche Spalten: {', '.join(missing_columns)}")
```

```python
def validate_and_convert_column(df, column_name, max_value=None):
    """
    Validates and converts the values in a specified column of a DataFrame to integers if they are numeric.
    Also checks that they do not exceed a given maximum value if provided.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the data to validate and convert.
    column_name (str): The name of the column to validate and convert.
    max_value (int, optional): The maximum valid value for the column; if None, no maximum check is performed.

    Returns:
    pd.DataFrame: DataFrame with the validated and converted column values.

    Raises:
    ValueError: If any value in the specified column cannot be converted to an integer or exceeds the maximum value, if provided.
    """
    if column_name not in df.columns:
        raise ValueError(f"The specified column '{column_name}' does not exist in the DataFrame.")

    try:
        # Attempt to convert all entries to integers
        df[column_name] = df[column_name].apply(lambda x: int(x) if pd.notna(x) and (isinstance(x, (int, float))) else x)

        # Validate entries
        if max_value is not None:
            if not df[column_name].apply(lambda x: x <= max_value if isinstance(x, int) else False).all():
                raise ValueError(f"Some entries in column '{column_name}' exceed the maximum value of {max_value}.")
        if not df[column_name].apply(lambda x: isinstance(x, int)).all():
            raise ValueError(f"Some entries in column '{column_name}' could not be converted to integers.")

        return df

    except Exception as e:
        raise ValueError(f"Error processing column '{column_name}': {e}")
```

```python
def preprocess_energy_data(df):
    """
    Führt die gesamte Vorverarbeitung und Validierung für den gegebenen DataFrame durch.
    """
    try:
        # Leere Datensätze entfernen
        df.dropna(inplace=True)

        # Erforderliche Spalten überprüfen
        required_columns = ['Jahr', 'Rubrik', 'Energietraeger', 'TJ']
        check_required_columns(df, required_columns)

        # Konvertierung und Validierung
        df['Rubrik'] = df['Rubrik'].str.replace('–', '-', regex=False)
        df = clean_tj_column(df)
        df = clean_and_validate_years(df)

        print("Vorverarbeitung und Validierung erfolgreich abgeschlossen.")
        return df

    except ValueError as e:
        print(f"Fehler während der Vorverarbeitung: {e}")
        return None

Energiedaten = preprocess_energy_data(Energiedaten)
if Energiedaten is not None:
  print(Energiedaten)
```

Daten zu sinnvollen Datasets zusammenfügen?

```python
def summarize_energy_data_per_year_per_column(df, category_column):
    """
    Summiert Energieverbrauchsdaten nach Jahr und einer spezifizierten Kategorie und berechnet die Prozentsätze jeder Kategorie.

    Parameters:
    df (pd.DataFrame): Der DataFrame, der die Energieverbrauchsdaten enthält.
    category_column (str): Der Name der Spalte, die die Kategorien beschreibt, nach denen summiert wird.

    Returns:
    pd.DataFrame: Ein DataFrame mit den summierten Daten und den berechneten Prozentsätzen für jede Kategorie.
    """
    # Gruppiere Daten nach Jahr und der spezifizierten Kategorie, und summiere die Energieverbrauchsdaten ('TJ') für jede Gruppe
    category_data_summed_per_year = df.groupby(['Jahr', category_column])['TJ'].sum().reset_index()

    # Gruppiere Daten nur nach Jahr und summiere die gesamte Energie ('TJ') für jedes Jahr
    total_energy_per_year = df.groupby('Jahr')['TJ'].sum().reset_index()
    total_energy_per_year.rename(columns={'TJ': 'TotalEnergy'}, inplace=True)  # Benenne die Spalte 'TJ' um in 'TotalEnergy'

    # Führe die Daten zu summierten Kategorien mit den totalen Energie-Daten zusammen, basierend auf dem Jahr
    summarized_data = pd.merge(category_data_summed_per_year, total_energy_per_year, on='Jahr')

    # Berechne den Prozentsatz des Verbrauchs jeder Kategorie relativ zur gesamten Energieverbrauchsmenge des jeweiligen Jahres
    summarized_data['Percentage'] = (summarized_data['TJ'] / summarized_data['TotalEnergy']) * 100

    return summarized_data
```

```python
def plot_pie(year):
    # Filter data for the selected year
    filtered_data = EnergieDatenEndverbrauchSummiert[EnergieDatenEndverbrauchSummiert['Jahr'] == year]
    labels = filtered_data['Energietraeger']
    sizes = filtered_data['Percentage']

    # Colors will be assigned by a colormap
    cmap = plt.get_cmap('tab20')  # Enhanced color map
    colors = [cmap(i) for i in np.linspace(0, 1, len(labels))]

    # Exploding the largest piece a little
    explode = [0.1 for i in sizes.index]

    # Create pie chart
    plt.figure(figsize=(10, 8))
    patches, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode, pctdistance=1.1)

    # Improve the display of the labels and percentage values
    for text, autotext in zip(texts, autotexts):
        text.set_fontsize(10)
        text.set_color('white')
        autotext.set_fontsize(10)

    # Add a legend
    plt.legend(patches, labels, title="Energieträger", loc="best", bbox_to_anchor=(1, 0, 0.5, 1))
    plt.title(f'Energieverbrauch nach Energieträger {year}')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()

plot_pie(2022)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
person_df = pd.read_csv('https://raw.githubusercontent.com/michaelzimmermannstudhslu/BINA_2024_energy_research/main/data/bip_total.csv')
energy_df = pd.read_csv('https://raw.githubusercontent.com/michaelzimmermannstudhslu/BINA_2024_energy_research/main/data/energiebilanz_ch.csv')

# Clean and process energy data
energy_df.dropna(inplace=True)
energy_df['TJ'] = pd.to_numeric(energy_df['TJ'], errors='coerce')
energy_df['TJ'] = energy_df['TJ'].abs()
energy_df.dropna(subset=['TJ'], inplace=True)
energy_df['TJ'] = energy_df['TJ'].astype(int)
energy_df = energy_df[energy_df['Rubrik'] == 'Endverbrauch - Total']

# Summarize energy data by year
total_energy_by_year = energy_df.groupby('Jahr')['TJ'].sum().reset_index()
total_energy_by_year.rename(columns={'TJ': 'TotalEnergy'}, inplace=True)

# Merge energy data with population data
final_merged_df = pd.merge(total_energy_by_year, person_df, on='Jahr')
final_merged_df['Energy_per_head'] = final_merged_df['TotalEnergy'] / final_merged_df['Bevölkerung']
final_merged_df['BIP_per_head'] = final_merged_df['BIP'] / final_merged_df['Bevölkerung']

# Calculate the annual change in BIP_per_head to find turning points
final_merged_df['BIP_change'] = final_merged_df['BIP_per_head'].pct_change()

# Identify turning points where the change in BIP_per_head changes sign
turning_points = final_merged_df[(final_merged_df['BIP_change'].shift(1) * final_merged_df['BIP_change'] < 0)]

# Create the plot
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot BIP per Head
color = 'tab:blue'
ax1.plot(final_merged_df['Jahr'], final_merged_df['BIP_per_head'], color=color, label='BIP_per_head')
ax1.set_xlabel('Jahr')
ax1.set_ylabel('BIP per Head', color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Mark turning points
ax1.plot(turning_points['Jahr'], turning_points['BIP_per_head'], 'go')  # 'go' for green circles

x_values = [final_merged_df['Jahr'].values[0], turning_points['Jahr'].values[0]]
y_values = [final_merged_df['BIP_per_head'].values[0], turning_points['BIP_per_head'].values[0]]
ax1.plot(x_values, y_values, 'g--')  # Green dashed line between points

# Draw lines between turning points
for i in range(len(turning_points['Jahr']) - 1):
    x_values = [turning_points['Jahr'].values[i], turning_points['Jahr'].values[i + 1]]
    y_values = [turning_points['BIP_per_head'].values[i], turning_points['BIP_per_head'].values[i + 1]]
    ax1.plot(x_values, y_values, 'g--')  # Green dashed line between points

# Plot Population on the primary y-axis
ax2 = ax1.twinx()
color = 'red'  # Set the color variable to red
ax2.plot(final_merged_df['Jahr'], final_merged_df['Bevölkerung'], color=color, label='Bevölkerung')  # Use 'color' variable
ax2.set_xlabel('Jahr')
ax2.set_ylabel('Bevölkerung', color=color)  # Set label color to red
ax2.tick_params(axis='y', labelcolor=color)  # Set tick color to red


# Title and legend
plt.title('Bevölkerung und BIP pro Kopf der Schweiz mit Wendepunkten (1980 - 2022)')
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/michaelzimmermannstudhslu/BINA_2024_energy_research/main/data/bip_total.csv')
df = df.dropna()

# Initialize the linear regression model
model_population = LinearRegression()
model_gdp = LinearRegression()

# Prepare data for regression (reshape year data for fitting)
X = df['Jahr'].values.reshape(-1, 1)  # Features
y_population = df['Bevölkerung'].values  # Target for population
y_gdp = df['BIP'].values  # Target for GDP

# Fit the models
model_population.fit(X, y_population)
model_gdp.fit(X, y_gdp)

# Predict the values
y_population_pred = model_population.predict(X)
y_gdp_pred = model_gdp.predict(X)

fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot population data
color = 'tab:red'
ax1.set_xlabel('Jahr')
ax1.set_ylabel('Bevölkerung', color=color)
ax1.plot(df['Jahr'], df['Bevölkerung'], marker='o', color=color, label='Actual Bevölkerung')
ax1.plot(df['Jahr'], y_population_pred, color=color, linestyle='--', label='Regression Line - Bevölkerung')
ax1.tick_params(axis='y', labelcolor=color)

# Second y-axis for GDP
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('BIP', color=color)
ax2.plot(df['Jahr'], df['BIP'], marker='o', color=color, label='Actual BIP')
ax2.plot(df['Jahr'], y_gdp_pred, color=color, linestyle='--', label='Regression Line - BIP')
ax2.tick_params(axis='y', labelcolor=color)

# Title and legend
plt.title('Bevölkerung und BIP der Schweiz mit Regression (1980 - 2022)')
fig.tight_layout()

# Handling legends for both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left')
ax2.legend(lines + lines2, labels + labels2, loc='upper left')
plt.show()
```
