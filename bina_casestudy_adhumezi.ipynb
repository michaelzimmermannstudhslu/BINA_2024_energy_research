```python
import pandas as pd
from collections import namedtuple
from pandas_schema import Column, Schema
from pandas_schema.validation import CanConvertValidation, InRangeValidation, InListValidation
```

```python
# Definiere eine Konstante für den Basis-URL-Pfad zu den GitHub-Daten
GITHUB_PATH = 'https://raw.githubusercontent.com/michaelzimmermannstudhslu/BINA_2024_energy_research/main/data/'

# Definiere einen benannten Tupeltyp für die Speicherung von Dateiinformationen
FileInfo = namedtuple('FileInfo', ['file_name', 'expected_columns'])

# Erstelle Instanzen von FileInfo für verschiedene Dateien, die die Dateinamen und die erwarteten Spalten enthalten
energiedaten_file_info = FileInfo(file_name='energiebilanz_ch.csv', expected_columns=['Jahr', 'Rubrik', 'Energietraeger'])
bevoelkerungBIP_file_info = FileInfo(file_name='bip_total.csv', expected_columns=['Jahr'])
CO2Ausstoss_file_info = FileInfo(file_name='co2.csv', expected_columns=['Jahr'])
CO2Ausstoss_Sektoren_file_info = FileInfo(file_name='co2_sector.csv', expected_columns=['Jahr'])

def load_and_inspect_data(file_info: FileInfo):
    """
    Lädt Daten von einer spezifizierten URL und bietet eine detaillierte Überprüfung des DataFrame.
    Optional werden auch eindeutige Werte in bestimmten Spalten überprüft.

    Parameter:
    - file_info (namedtuple): Enthält den Dateinamen und die erwarteten Spalten.
    - columns_to_check (list of str, optional): Liste von Spaltennamen, für die eindeutige Werte überprüft werden sollen.

    Rückgabe:
    - DataFrame: Der geladene und inspizierte DataFrame, falls erfolgreich.
    - None: Gibt None zurück, wenn beim Laden oder Verarbeiten der Daten Probleme auftraten.
    """
    url = GITHUB_PATH + file_info.file_name  # Konstruiere die vollständige URL durch Kombination der Basis-URL mit dem Dateinamen

    try:
        # Lade Daten von der vollständigen URL
        data = pd.read_csv(url)

        # Zeige Informationen über den DataFrame an, wie Datentypen und Anzahl der Zeilen/Spalten
        print("Informationen über den DataFrame:")
        print(data.info())

        # Zeige die ersten fünf Zeilen des DataFrames, um einen Eindruck von den Daten zu erhalten
        print("Erste fünf Zeilen des DataFrames:")
        print(data.head())

        # Zeige die letzten fünf Zeilen des DataFrames, um die Daten am Ende des Datensatzes zu sehen
        print("Letzte fünf Zeilen des DataFrames:")
        print(data.tail())

        # Überprüfe und zeige eindeutige Werte für die spezifizierten Spalten, falls angegeben
        if file_info.expected_columns is not None:
            for column in file_info.expected_columns:
                if column not in data.columns:
                    raise ValueError(f"Spalte '{column}' nicht im DataFrame gefunden.")
                print(f"Einzigartige Werte in der Spalte '{column}':")
                print(data[column].unique())

        return data

    except pd.errors.EmptyDataError:
        print("Keine Daten gefunden: Die Datei ist leer.")
        return None

    except pd.errors.ParserError:
        print("Fehler beim Parsen der Daten: Überprüfen Sie das Format der CSV-Datei.")
        return None

    except FileNotFoundError:
        print("Datei nicht gefunden: Überprüfen Sie den Dateipfad oder die URL.")
        return None

    except Exception as e:
        print(f"Ein unerwarteter Fehler ist aufgetreten: {e}")
        return None
```

```python
# Die Daten des Endverbrauchs sind nicht ganz richtig. Der Endverbrauch - Total in Bezug der Energieträger umfasst: Elektrizität, Erdölprodukte, Fernwärme, Gas, Holzenergie, Kohle, Müll und Industrieabfälle und Übrige erneuerbare Energien.
# Hier müssen Kernbrennstoffe, Rohöl und Wasserkraft entfernt werden

# Erlaubte Werte für 'Rubrik' definieren
allowed_rubriken = [
    'Bruttoverbrauch',
    'Eigenverbrauch des Energiesektors, Netzverluste, Verbrauch der Speicherungen',
    'Endverbrauch - Total',
    'Export', 'Import', 'Inlandproduktion', 'Lagerveränderung',
    'Nichtenergetischer Verbrauch'
]

enverbrauch_rubriken = [
    'Endverbrauch - Dienstleistungen', 'Endverbrauch - Haushalte',
    'Endverbrauch - Industrie',
    'Endverbrauch - Statistische Differenz inkl. Landwirtschaft',
    'Endverbrauch - Verkehr'
]

# Erlaubte Werte für 'Energieträger' definieren, aufgeteilt in erneuerbare und nicht-erneuerbare Energiequellen
renewable_rubriken = [
    'Energieumwandlung - Wasserkraftwerke', 'Energieumwandlung - Diverse Erneuerbare'
]

non_renewable_rubriken = [
    'Energieumwandlung - Gaswerke', 'Energieumwandlung - Kernkraftwerke', 'Energieumwandlung - Raffinerien', 'Energieumwandlung - konventionell-thermische Kraft-, Fernheiz- und Fernheizkraftwerke',
]

# Erlaubte Werte für 'Energieträger' definieren, aufgeteilt in erneuerbare und nicht-erneuerbare Energiequellen
renewable_energietraeger = [
    'Wasserkraft', 'Uebrige erneuerbare Energien', 'Holzenergie'
]

non_renewable_energietraeger = [
    'Elektrizität', 'Erdölprodukte', 'Fernwärme', 'Gas',
    'Kernbrennstoffe', 'Kohle', 'Müll und Industrieabfälle', 'Rohöl'
]

# Definition des Schemas für Daten über das Bruttoinlandsprodukt und Bevölkerung
BIP_SCHEMA = Schema([
    Column('Jahr', [CanConvertValidation(int), InRangeValidation(1980, 2024)]),
    Column('Bevölkerung', [CanConvertValidation(int), InRangeValidation(0, 1000000000)]),
    Column('BIP', [CanConvertValidation(float), InRangeValidation(0, float('inf'))]),
    Column('Bip_pro_Kopf', [CanConvertValidation(float), InRangeValidation(0, float('inf'))])
])

# Definition des Schemas für das Energie-Daten-DataFrame mit Überprüfung auf die erlaubten Rubriken und Energieträger
ENERGY_SCHEMA = Schema([
    Column('Jahr', [CanConvertValidation(int), InRangeValidation(1980, 2024)]),
    Column('Rubrik', [CanConvertValidation(str), InListValidation(allowed_rubriken + enverbrauch_rubriken + renewable_rubriken + non_renewable_rubriken)]),
    Column('Energietraeger', [CanConvertValidation(str), InListValidation(renewable_energietraeger + non_renewable_energietraeger)]),
    Column('TJ', [CanConvertValidation(float)])
])

# Definition des Schemas für CO2-Emissionsdaten
CO2_SCHEMA = Schema([
    Column('Jahr', [CanConvertValidation(int), InRangeValidation(1980, 2024)]),
    Column('THG', [CanConvertValidation(float)])
])

# Definition des Schemas für CO2-Emissionsdaten über verschiedene Sektoren hinweg
CO2_SECTOR_SCHEMA = Schema([
    Column('Jahr', [CanConvertValidation(int), InRangeValidation(1980, 2024)]),
    Column('Dienstleistungen', [CanConvertValidation(float), InRangeValidation(0, float('inf'))]),
    Column('Haushalte', [CanConvertValidation(float), InRangeValidation(0, float('inf'))]),
    Column('Verkehr', [CanConvertValidation(float), InRangeValidation(0, float('inf'))]),
    Column('Industrie', [CanConvertValidation(float), InRangeValidation(0, float('inf'))]),
    Column('Übrige', [CanConvertValidation(float), InRangeValidation(0, float('inf'))])
])

def validate_and_clean_dataframe(df: pd.DataFrame, schema: Schema):
    """
    Überprüft und bereinigt einen DataFrame basierend auf einem gegebenen Schema. Entfernt Spalten, die nicht im Schema definiert sind.

    Parameter:
    - df (pd.DataFrame): Der zu validierende DataFrame.
    - schema (Schema): Das Schema, gegen das der DataFrame validiert wird.

    Rückgabe:
    - df (pd.DataFrame): Der bereinigte DataFrame, der nur die im Schema definierten Spalten enthält.
    """
    schema_columns = [column.name for column in schema.columns]  # Extrahiere Spaltennamen aus dem Schema
    df = df.loc[:, schema_columns]  # Behalte nur die Spalten, die im Schema definiert sind
    return df

def validate_dataframe(df: pd.DataFrame, schema: Schema):
    """
    Validiert einen DataFrame gegen ein bereitgestelltes Schema. Wirft eine Ausnahme, wenn Validierungsfehler gefunden werden.

    Parameter:
    - df (pd.DataFrame): Der zu validierende DataFrame.
    - schema (Schema): Das Schema, gegen das der DataFrame validiert wird.

    Fehler:
    - ValueError: Wird ausgelöst, wenn im DataFrame Validierungsfehler gefunden werden.

    Rückgabe:
    - None
    """
    errors = schema.validate(df)  # Validiere den DataFrame mit dem bereitgestellten Schema
    if errors:
        error_messages = "\n".join(str(e) for e in errors)  # Sammle alle Fehlermeldungen in einem String
        print(f"Validation errors occurred:\n{error_messages}")  # Drucke Fehler zur Protokollierung
        raise ValueError(f"Validation errors occurred:\n{error_messages}")  # Wirf eine Ausnahme mit allen Fehlermeldungen
    else:
        print("Alle Daten sind gültig.")  # Bestätigung, dass alle Daten gültig sind, wenn keine Fehler gefunden wurden

def report_nan_values(df):
  """
  Erstellt einen Bericht über NaN-Werte in einem DataFrame und gibt den gesamten Eintrag für jede Zeile mit NaN-Werten aus.

  Parameter:
  df (pd.DataFrame): Der DataFrame, der überprüft werden soll.

  Rückgabe:
  None
  """
  nan_rows = df[df.isnull().any(axis=1)]

  if len(nan_rows) == 0:
      print("Der DataFrame enthält keine NaN-Werte.")
  else:
      print("Der DataFrame enthält NaN-Werte. Hier sind die betroffenen Zeilen:")
      for index, row in nan_rows.iterrows():
          print(row)
```

### 3.1.3 Daten organisieren

In diesem Abschnitt werden die vorbereiteten Daten organisiert. Dabei werden sinnvolle Datasets zusammengestellt, damit die entsprechenden Daten in notwendigen Qualität und Quantität für die Analyse und die Präsentation zur Verfügung stehen und effizient verwendet werden können.


```python
def summarize_energy_data_per_year_per_column(df, category_column):
    """
    Summiert die Energieverbrauchsdaten pro Jahr und nach der spezifizierten Kategorie.
    Berechnet auch die prozentuale Verteilung jeder Kategorie im Verhältnis zum Gesamtverbrauch pro Jahr.

    Parameter:
    df (pd.DataFrame): Der DataFrame, der die Energieverbrauchsdaten enthält.
    category_column (str): Der Name der Spalte, die die zu summierenden Kategorien angibt.

    Rückgabe:
    pd.DataFrame: Ein DataFrame, der die summierten Daten und die berechneten Prozentsätze für jede Kategorie enthält.
    """
    try:
        # Konvertiert die 'TJ' Spalte in numerische Werte und behandelt Fehler durch Umwandlung nicht-numerischer Werte in NaN
        df['TJ'] = pd.to_numeric(df['TJ'], errors='coerce')

        # Entfernt Zeilen, in denen 'TJ' NaN ist, was nach der Konvertierung passieren könnte
        df.dropna(subset=['TJ'], inplace=True)

        # Gruppierung der Daten nach 'Jahr' und der spezifizierten Kategorie und Summierung der 'TJ'-Werte für jede Gruppe
        category_data_summed_per_year = df.groupby(['Jahr', category_column])['TJ'].sum().reset_index()

        # Summierung der gesamten Energie ('TJ') nur nach 'Jahr'
        total_energy_per_year = df.groupby('Jahr')['TJ'].sum().reset_index()

        # Umbenennen der Spalte 'TJ' in 'TotalEnergy' für eine klarere Darstellung
        total_energy_per_year.rename(columns={'TJ': 'TotalEnergy'}, inplace=True)

        # Zusammenführen der nach Kategorien summierten Daten mit den gesamten Energie-Daten nach 'Jahr'
        summarized_data = pd.merge(category_data_summed_per_year, total_energy_per_year, on='Jahr')

        # Berechnung des prozentualen Anteils des Verbrauchs jeder Kategorie im Vergleich zum Gesamtenergieverbrauch des jeweiligen Jahres
        summarized_data['Percentage'] = (summarized_data['TJ'] / summarized_data['TotalEnergy']) * 100

        # Rückgabe des bearbeiteten DataFrames, der sowohl die absoluten Werte als auch die Prozentsätze enthält
        return summarized_data
    except Exception as e:
        raise ValueError("Ein Fehler ist beim Zusammenfassen der Daten aufgetreten: {}".format(e))
```

Nachfolgend werden Funktionen definiert, um die Diagramme effizienter zu bearbeiten und darzustellen.


```python
# Defining colors that make sense to the words
assosiated_colors = {
    'Endverbrauch - Dienstleistungen': 'blue',
    'Endverbrauch - Haushalte': 'green',
    'Endverbrauch - Industrie': 'orange',
    'Endverbrauch - Statistische Differenz inkl. Landwirtschaft': 'red',
    'Endverbrauch - Verkehr': 'purple',
    'Energieumwandlung - Wasserkraftwerke': 'lightblue',
    'Energieumwandlung - Diverse Erneuerbare': 'green',
    'Energieumwandlung - Gaswerke': 'orange',
    'Energieumwandlung - Kernkraftwerke': 'red',
    'Energieumwandlung - Raffinerien': 'purple',
    'Energieumwandlung - konventionell-thermische Kraft-, Fernheiz- und Fernheizkraftwerke': 'brown',
    'Wasserkraft': 'lightblue',
    'Uebrige erneuerbare Energien': 'green',
    'Holzenergie': 'brown',
    'Elektrizität': 'yellow',
    'Erdölprodukte': 'orange',
    'Fernwärme': 'red',
    'Gas': 'purple',
    'Kernbrennstoffe': 'brown',
    'Kohle': 'black',
    'Müll und Industrieabfälle': 'gray',
    'Rohöl': 'darkorange'
}

def darken_color(color, amount=0.2):
    """
    Verdunkelt die angegebene Farbe, indem ihre Helligkeit oder Leuchtkraft reduziert wird.

    Parameter:
    - color (str oder tuple): Die zu verdunkelnde Farbe.
    - amount (float): Das Maß, um das die Farbe verdunkelt werden soll (0,0 bis 1,0).

    Rückgabe:
    - str: Die verdunkelte Farbe im hexadezimalen Format.
    """
    try:
        c = mcolors.hex2color(color)
    except ValueError:
        c = mcolors.to_rgb(color)
    return mcolors.to_hex(tuple(max(0, min(1, channel - amount)) for channel in c))

def plot_stacked_area(data, x_label, y_label, title, legend_title, trendline_per_column=False, total_trendline=False, turning_points=False):
    """
    Erstellt ein gestapeltes Flächendiagramm.

    Parameter:
    - data (DataFrame): DataFrame, der die Daten für das Diagramm enthält, indiziert durch die Werte der x-Achse.
    - x_label (str): Beschriftung der x-Achse.
    - y_label (str): Beschriftung der y-Achse.
    - title (str): Titel des Diagramms.
    - legend_title (str): Titel für die Legende.
    - trendline_per_column (bool): Flag, um Trendlinien für jede Spalte zu aktivieren oder zu deaktivieren (Standard: False).
    - total_trendline (bool): Flag, um die Gesamttrendlinie zu aktivieren oder zu deaktivieren (Standard: False).
    - turning_points (bool): Flag, um die Wendepunkte zu aktivieren oder zu deaktivieren (Standard: False).
    """
    # Check if all columns are in assosiated_colors
    for col in data.columns:
        if col not in assosiated_colors:
            raise ValueError(f"Color for '{col}' is not defined in assosiated_colors.")

    plt.figure(figsize=(12, 8))
    plt.stackplot(data.index, data.values.T, labels=data.columns, colors=[assosiated_colors[col] for col in data.columns])

    # Trendlinien plotten und lineare Regression für jede Spalte berechnen
    if trendline_per_column:
        for i, col in enumerate(data.columns):
            x = data.index
            y = data[col]
            slope, intercept, _, _, _ = linregress(x, y)
            trend_line = slope * x + intercept
            plt.plot(x, trend_line, color=darken_color(assosiated_colors[col]), alpha=1, linestyle='--', linewidth=2.0, label=f'Trendlinie ({col})')

    # Summe der Werte entlang der Y-Achse
    y_total = data.sum(axis=1)

    # Trendlinie für das gesamte Y berechnen
    if total_trendline:
        x_total = data.index
        slope_total, intercept_total, _, _, _ = linregress(x_total, y_total)
        trend_line_total = slope_total * x_total + intercept_total
        plt.plot(x_total, trend_line_total, color='black', alpha=1, linestyle='--', linewidth=2.0, label=f'Trendlinie (Gesamt)')

    # Wendepunkte für das gesamte Y finden
    if turning_points:
        second_derivative_total = np.gradient(np.gradient(y_total, x_total), x_total)
        turning_points_total = []
        for j in range(1, len(second_derivative_total) - 1):  # Ersten und letzten Punkt ausschließen
            if np.sign(second_derivative_total[j-1]) != np.sign(second_derivative_total[j]):
                turning_points_total.append(x_total[j])
        plt.scatter(turning_points_total, y_total[np.isin(x_total, turning_points_total)], color='red', marker='x', zorder=5)

        # Vertikale Linien an Wendepunkten für das Gesamt-Y zeichnen
        for tp in turning_points_total:
            plt.vlines(tp, 0, y_total.max(), color='black', linestyle='--', linewidth=1)

    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title=legend_title)
    return plt

def plot_stacked_bar(data, title, xlabel, ylabel, legend_title='Energiearten', figsize=(12, 8), trendline_per_column=False, total_trendline=False, turning_points=False):
    """
    Erstellt ein gestapeltes Balkendiagramm mit gegebenen Daten.

    Parameter:
    - data (DataFrame): Ein pandas DataFrame, wobei jede Spalte eine unterschiedliche Kategorie darstellt und der Index die x-Achse repräsentiert.
    - title (str): Der Titel des Diagramms.
    - xlabel (str): Beschriftung der x-Achse.
    - ylabel (str): Beschriftung der y-Achse.
    - legend_title (str): Titel für die Legende.
    - figsize (tuple): Tupel für die Größe der Figur (Breite, Höhe).
    """
    # Check if all columns are in assosiated_colors
    for col in data.columns:
        if col not in assosiated_colors:
            raise ValueError(f"Color for '{col}' is not defined in assosiated_colors.")

    fig, ax = plt.subplots(figsize=figsize)
    bottom = np.zeros(len(data))  # Dies wird die kumulative Untergrenze für jede Stapelung verfolgen

    # Jede Spalte als separater Teil des Stapels plotten, Farben zuteilen
    for idx, col in enumerate(data.columns):
        ax.bar(data.index, data[col], bottom=bottom, label=col, color=assosiated_colors[col])
        bottom += data[col].values  # Aktualisieren der Untergrenze für den nächsten Stapel

    # Trendlinien plotten und lineare Regression für jede Spalte berechnen
    if trendline_per_column:
        for i, col in enumerate(data.columns):
          # Gleitenden Durchschnitt berechnen und plotten
          rolling_mean = data[col].rolling(window=5, min_periods=1).mean()
          ax.plot(data.index, rolling_mean, linestyle='--', linewidth=2.0, color=darken_color(assosiated_colors[col]), label=f'Gleitender Trend ({col})')

    # Summe der Werte entlang der Y-Achse
    y_total = data.sum(axis=1)

    # Trendlinie für das gesamte Y berechnen
    if total_trendline:
        x_total = data.index
        slope_total, intercept_total, _, _, _ = linregress(x_total, y_total)
        trend_line_total = slope_total * x_total + intercept_total
        ax.plot(x_total, trend_line_total, color='black', alpha=1, linestyle='--', linewidth=2.0, label=f'Trendlinie (Gesamt)')

    # Wendepunkte für das gesamte Y finden
    if turning_points:
        second_derivative_total = np.gradient(np.gradient(y_total, x_total), x_total)
        turning_points_total = []
        for j in range(1, len(second_derivative_total) - 1):  # Ersten und letzten Punkt ausschließen
            if np.sign(second_derivative_total[j-1]) != np.sign(second_derivative_total[j]):
                turning_points_total.append(x_total[j])
        ax.scatter(turning_points_total, y_total[np.isin(x_total, turning_points_total)], color='red', marker='x', zorder=5)

        # Vertikale Linien an Wendepunkten für das Gesamt-Y zeichnen
        for tp in turning_points_total:
            ax.vlines(tp, 0, y_total.max(), color='black', linestyle='--', linewidth=1)

    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(loc='upper left', bbox_to_anchor=(1, 1), title=legend_title)
    return plt

def plot_multiple_lines(data, title, xlabel, ylabel, legend_title='Energiearten', marker='o', figsize=(12, 8)):
    """
    Zeichnet mehrere Linien auf einem einzigen Diagramm.

    Parameter:
    - data (DataFrame): Ein pandas DataFrame, wobei jede Spalte eine unterschiedliche Linie darstellt und der Index die x-Achse repräsentiert.
    - title (str): Der Titel des Diagramms.
    - xlabel (str): Beschriftung der x-Achse.
    - ylabel (str): Beschriftung der y-Achse.
    - legend_title (str): Titel für die Legende.
    - marker (str): Markierungsstil für die Datenpunkte auf den Linien.
    - figsize (tuple): Tupel für die Größe der Figur (Breite, Höhe).
    """
    # Check if all columns are in assosiated_colors
    for col in data.columns:
        if col not in assosiated_colors:
            raise ValueError(f"Color for '{col}' is not defined in assosiated_colors.")

    fig, ax = plt.subplots(figsize=figsize)

    # Jede Spalte als separate Linie plotten
    for col in data.columns:
        color = assosiated_colors[col]
        ax.plot(data.index, data[col], marker=marker, label=col, color=color)  # Farbe für die Linie setzen

        # Trendlinie plotten und lineare Regression berechnen
        x = data.index
        y = data[col]
        slope, intercept, _, _, _ = linregress(x, y)
        trend_line = slope * x + intercept
        first_color = darken_color(color)
        plt.plot(x, trend_line, alpha=1, linestyle='--', linewidth=2.0, color=first_color, label=f'Trendlinie ({col})')

        # Gleitenden Durchschnitt berechnen und plotten
        rolling_mean = data[col].rolling(window=5, min_periods=1).mean()
        ax.plot(data.index, rolling_mean, linestyle='--', linewidth=2.0, color=darken_color(first_color), label=f'Gleitender Trend ({col})')


    # Diagramm mit Titeln und Beschriftungen einrichten
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(title=legend_title, loc='upper left', bbox_to_anchor=(1, 1))

    # y-Achse beginnt bei 0
    ax.set_ylim(bottom=0)

    # Diagramm anzeigen
    return plt

def index_plot_multiple_lines(data, title, xlabel, ylabel, legend_title='Energiearten', marker='o', figsize=(12, 8)):
    """
    Zeichnet mehrere Linien auf einem einzigen Diagramm.

    Parameter:
    - data (DataFrame): Ein pandas DataFrame, wobei jede Spalte eine unterschiedliche Linie darstellt und der Index die x-Achse repräsentiert.
    - title (str): Der Titel des Diagramms.
    - xlabel (str): Beschriftung der x-Achse.
    - ylabel (str): Beschriftung der y-Achse.
    - legend_title (str): Titel für die Legende.
    - marker (str): Markierungsstil für die Datenpunkte auf den Linien.
    - figsize (tuple): Tupel für die Größe der Figur (Breite, Höhe).
    """
    # Check if all columns are in assosiated_colors
    for col in data.columns:
        if col not in assosiated_colors:
            raise ValueError(f"Color for '{col}' is not defined in assosiated_colors.")

    # Normalize the data so each column starts at 100
    def normalize_series(series):
        if series.iloc[0] == 0:
            series.iloc[0] = 100
        return (series / series.iloc[0]) * 100

    normalized_data = data.apply(normalize_series)

    fig, ax = plt.subplots(figsize=figsize)

    # Jede Spalte als separate Linie plotten
    for col in normalized_data.columns:
        color = assosiated_colors[col]
        ax.plot(normalized_data.index, normalized_data[col], marker=marker, label=col, color=color)  # Farbe für die Linie setzen

    # Diagramm mit Titeln und Beschriftungen einrichten
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(title=legend_title, loc='upper left', bbox_to_anchor=(1, 1))

    # y-Achse beginnt bei 0
    ax.set_ylim(bottom=0)

    # Diagramm anzeigen
    return plt

def index_mean_plot_multiple_lines(data, title, xlabel, ylabel, legend_title='Energiearten', marker='o', figsize=(12, 8)):
    """
    Zeichnet mehrere Linien auf einem einzigen Diagramm.

    Parameter:
    - data (DataFrame): Ein pandas DataFrame, wobei jede Spalte eine unterschiedliche Linie darstellt und der Index die x-Achse repräsentiert.
    - title (str): Der Titel des Diagramms.
    - xlabel (str): Beschriftung der x-Achse.
    - ylabel (str): Beschriftung der y-Achse.
    - legend_title (str): Titel für die Legende.
    - marker (str): Markierungsstil für die Datenpunkte auf den Linien.
    - figsize (tuple): Tupel für die Größe der Figur (Breite, Höhe).
    """
    # Check if all columns are in assosiated_colors
    for col in data.columns:
        if col not in assosiated_colors:
            raise ValueError(f"Color for '{col}' is not defined in assosiated_colors.")

    # Normalize the data so each column starts at 100
    normalized_data = data.apply(lambda x: (x / x.iloc[0]) * 100)

    fig, ax = plt.subplots(figsize=figsize)

    # Jede Spalte als separate Linie plotten
    for col in normalized_data.columns:
        color = assosiated_colors[col]
        # Gleitenden Durchschnitt berechnen und plotten
        rolling_mean = normalized_data[col].rolling(window=5, min_periods=1).mean()
        ax.plot(normalized_data.index, rolling_mean, linestyle='--', linewidth=2.0, color=color, label=f'Gleitender Trend ({col})')

    # Diagramm mit Titeln und Beschriftungen einrichten
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(title=legend_title, loc='upper left', bbox_to_anchor=(1, 1))

    # y-Achse beginnt bei 0
    ax.set_ylim(bottom=0)

    # Diagramm anzeigen
    return plt



def plot_horizontal_bar(df, year, carrier_col='Energietraeger', percentage_col='Percentage', plot_title=None):
    """
    Erstellt ein horizontales Balkendiagramm des Energieverbrauchs nach bestimmten Energieträgern für ein vorgegebenes Jahr aus einem bereitgestellten DataFrame.

    Parameter:
    - df (DataFrame): Der DataFrame, der die Energiedaten enthält.
    - year (int): Das Jahr, für das die Daten visualisiert werden sollen.
    - carrier_col (str): Spaltenname im DataFrame, der die Energieträger kennzeichnet.
    - percentage_col (str): Spaltenname im DataFrame, der die Prozentsätze darstellt.
    - plot_title (str): Titel des Diagramms, falls None, wird ein Standardtitel generiert.

    Der DataFrame sollte mindestens die Spalten nach `carrier_col` und `percentage_col` neben 'Jahr' enthalten.
    """
    # Daten für das ausgewählte Jahr filtern
    filtered_data = df[df['Jahr'] == year]
    labels = filtered_data[carrier_col]
    sizes = filtered_data[percentage_col]

    # Überprüfen, ob alle Labels im assosiated_colors Wörterbuch vorhanden sind
    for label in labels:
        if label not in assosiated_colors:
            raise ValueError(f"Farbe für '{label}' ist nicht im assosiated_colors Wörterbuch definiert.")

    # Sortiere die Daten nach Größen in absteigender Reihenfolge für eine bessere Visualisierung
    sorted_indices = sizes.argsort()
    sizes = sizes.iloc[sorted_indices]
    labels = labels.iloc[sorted_indices]

    # Farben werden aus dem assosiated_colors Wörterbuch zugewiesen
    colors = [assosiated_colors[label] for label in labels]

    # Erstelle ein horizontales Balkendiagramm
    plt.figure(figsize=(12, 8))
    bars = plt.barh(labels, sizes, color=colors, edgecolor='black')

    # Füge Wertbeschriftungen zu jedem Balken hinzu
    for bar in bars:
        plt.text(bar.get_width() + 1,  # Ein kleines Offset hinzufügen, um den Text außerhalb des Balkens zu platzieren
                 bar.get_y() + bar.get_height()/2,
                 f'{bar.get_width():.1f}%',
                 va='center', ha='left', color='black', fontsize=10, fontweight='bold')

    # Titel und Beschriftungen
    if plot_title is None:
        plot_title = f'Energieverbrauch nach Energieträger im Jahr {year}'
    plt.title(plot_title)
    plt.xlabel('Prozentanteil (%)')
    plt.ylabel('Energieträger')

    # Setze die x-Achse auf einen Bereich von 0 bis 60
    plt.xlim(0, 60)

    # Zeige das Diagramm an
    plt.gca().invert_yaxis()  # Invertiere die y-Achse, um das größte Element oben anzuzeigen
    plt.show()

def plot_treemap(df, year, carrier_col='Energietraeger', percentage_col='Percentage', plot_title=None):
    """
    Erstellt ein TreeMap-Diagramm des Energieverbrauchs nach bestimmten Energieträgern für ein vorgegebenes Jahr aus einem bereitgestellten DataFrame.

    Parameter:
    - df (DataFrame): Der DataFrame, der die Energiedaten enthält.
    - year (int): Das Jahr, für das die Daten visualisiert werden sollen.
    - carrier_col (str): Spaltenname im DataFrame, der die Energieträger kennzeichnet.
    - percentage_col (str): Spaltenname im DataFrame, der die Prozentsätze darstellt.
    - plot_title (str): Titel des Diagramms, falls None, wird ein Standardtitel generiert.

    Der DataFrame sollte mindestens die Spalten nach `carrier_col` und `percentage_col` neben 'Jahr' enthalten.
    """
    # Daten für das ausgewählte Jahr filtern
    filtered_data = df[df['Jahr'] == year]
    filtered_data = filtered_data[filtered_data[percentage_col] != 0]
    labels = filtered_data[carrier_col]
    sizes = filtered_data[percentage_col]

    # Überprüfen, ob alle Labels im assosiated_colors Wörterbuch vorhanden sind
    for label in labels:
        if label not in assosiated_colors:
            raise ValueError(f"Farbe für '{label}' ist nicht im assosiated_colors Wörterbuch definiert.")

    # Sortiere die Daten nach Größe in absteigender Reihenfolge für eine bessere Visualisierung
    sorted_indices = sizes.argsort()
    sizes = sizes.iloc[sorted_indices]
    labels = labels.iloc[sorted_indices]

    # Farben werden aus dem assosiated_colors Wörterbuch zugewiesen
    colors = [assosiated_colors[label] for label in labels]

    # Erstelle das TreeMap
    plt.figure(figsize=(35, 12))
    ax = squarify.plot(sizes=sizes, label=labels, color=colors, alpha=0.8, text_kwargs={'fontsize': 12, 'fontweight': 'bold'})

    # Füge Prozentwerte neben den Labels hinzu
    for i, size in enumerate(sizes):
        if size > 0:
          x, y, dx, dy = ax.patches[i].get_x(), ax.patches[i].get_y(), ax.patches[i].get_width() + 12, ax.patches[i].get_height()
          plt.text(x + dx / 2, y + dy / 2, f'{size:.1f}%', horizontalalignment='center', verticalalignment='center', fontsize=10, fontweight='normal')

    # Titel setzen
    if plot_title is None:
        plot_title = f'Energieverbrauch nach Energieträger im Jahr {year}'
    plt.title(plot_title, fontsize=16)

    # Entferne Achsen
    plt.axis('off')

    # Zeige das TreeMap an
    plt.show()
```

```python
# Kommentar Andi 14.05.24
# Trendberechnung (via 'slope') der Sektoren durchführen und als Kommazahl (z.B. 0.0111.. / -0.9423...) zur Interpretation hier im Analyseteil ausgeben. Dabei soll der Trend aller Sektoren einzel aus Kommazahl ausgegeben werden.

# Deskriptive Statistiken nach Sektoren
grouped_data = EnergieDatenSektorenSummiert.groupby('Rubrik')
print(grouped_data['TJ'].describe())

def get_min_max_years_and_values(group):
    """
    Gibt die Jahre mit dem minimalen und maximalen Energieverbrauch sowie deren entsprechende Werte für die Gruppe zurück.
    """
    # Indizes des minimalen und maximalen Energieverbrauchs ermitteln
    min_idx = group['TJ'].idxmin()
    max_idx = group['TJ'].idxmax()

    # Eine Serie mit den minimalen und maximalen Jahren und Werten zurückgeben
    return pd.Series({
        'Min Year': group.loc[min_idx, 'Jahr'],
        'Max Year': group.loc[max_idx, 'Jahr'],
        'Min TJ': group.loc[min_idx, 'TJ'],
        'Max TJ': group.loc[max_idx, 'TJ']
    })

# Funktion auf jede Gruppe anwenden
min_max_years = grouped_data.apply(get_min_max_years_and_values)
print(min_max_years)
```

```python
# Plotting the area chart
fig, ax = plt.subplots(figsize=(10, 6))
CO2Ausstoss_Sektoren.plot(kind='area', stacked=True, ax=ax)

# Calculate cumulative sum for total contributions per year
total_contributions = CO2Ausstoss_Sektoren.sum(axis=1)

# Use numpy to calculate a smoothed version of the total contributions
smoothed = savgol_filter(total_contributions, 9, 3)  # Using Savitzky-Golay filter for better smoothing

# Calculate second derivative
second_derivative = np.diff(smoothed, n=2)

# Define a threshold for "big changes"
threshold = np.percentile(np.abs(second_derivative), 95)  # top 5% magnitude changes

# Identify inflection points (zero crossings in the second derivative)
inflection_points = np.where(np.diff(np.sign(second_derivative)))[0] + 1  # Adjust for shift caused by diff

# Plot inflection points
for point in inflection_points:
    if point < len(CO2Ausstoss_Sektoren) and np.abs(second_derivative[point - 1]) > threshold:
        year = CO2Ausstoss_Sektoren.index[point]
        ax.plot(year, total_contributions.loc[year], 'ro')  # Mark inflection point with a red circle
        ax.axvline(x=year, color='black', linestyle='--', linewidth=1)  # Draw vertical line

# Fit trend lines for specified periods
# Trend line for 1990-2011
years_1990_2011 = np.arange(1990, 2012)
contributions_1990_2011 = total_contributions.loc[1990:2011]
coeffs_1990_2011 = np.polyfit(years_1990_2011, contributions_1990_2011, 1)
poly_1990_2011 = np.poly1d(coeffs_1990_2011)
ax.plot(years_1990_2011, poly_1990_2011(years_1990_2011), 'b--', label='1990-2011 Trend')

# Trend line for 2011-2021
years_2011_2021 = np.arange(2011, 2022)
contributions_2011_2021 = total_contributions.loc[2011:2021]
coeffs_2011_2021 = np.polyfit(years_2011_2021, contributions_2011_2021, 1)
poly_2011_2021 = np.poly1d(coeffs_2011_2021)
ax.plot(years_2011_2021, poly_2011_2021(years_2011_2021), 'b--', label='2011-2021 Trend')

# Customize the plot
plt.title('Stacked Area Chart of Sector Contributions Over Time with Significant Inflection Points')
plt.xlabel('Year')
plt.ylabel('Contributions')
plt.tight_layout()
plt.show()
```

```python
# Prepare the data for regression
X = CO2Ausstoss[['Jahr']]  # Double brackets to ensure X is a DataFrame with one column
y = CO2Ausstoss['THG']  # Response variable

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Coefficients of the model
slope = model.coef_[0]
intercept = model.intercept_

# Predict when THG will be zero
# THG = intercept + slope * Jahr, set THG to 0 and solve for Jahr
if slope != 0:
    year_zero_emissions = int(-intercept / slope)
else:
    year_zero_emissions = "never" if intercept > 0 else "already at zero"

# Create a range of years for prediction
years_extended = np.arange(1990, year_zero_emissions + 1 if year_zero_emissions != "never" else 2050)
future_X = years_extended.reshape(-1, 1)

# Predict for extended years
predicted_THG = model.predict(future_X)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(CO2Ausstoss['Jahr'], CO2Ausstoss['THG'], color='blue', label='Historical THG Data')
plt.plot(years_extended, predicted_THG, color='red', linestyle='--', label='Trend Line')
plt.axhline(0, color='green', linestyle='-', label='Zero Emissions Line')
plt.axvline(x=year_zero_emissions, color='purple', linestyle='--', label=f'Estimated Year: {year_zero_emissions}')

plt.title('Predicted Year for Zero Total Greenhouse Gas Emissions')
plt.xlabel('Year')
plt.ylabel('Total Greenhouse Gas Emissions (THG)')
plt.legend()
plt.grid(True)
plt.show()

# Output the estimated year of zero emissions
print(f"Estimated year when THG will reach zero based on the trend: {year_zero_emissions}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
person_df = pd.read_csv('https://raw.githubusercontent.com/michaelzimmermannstudhslu/BINA_2024_energy_research/main/data/bip_total.csv')
energy_df = pd.read_csv('https://raw.githubusercontent.com/michaelzimmermannstudhslu/BINA_2024_energy_research/main/data/energiebilanz_ch.csv')

# Clean and process energy data
energy_df.dropna(inplace=True)
energy_df['TJ'] = pd.to_numeric(energy_df['TJ'], errors='coerce')
energy_df['TJ'] = energy_df['TJ'].abs()
energy_df.dropna(subset=['TJ'], inplace=True)
energy_df['TJ'] = energy_df['TJ'].astype(int)
energy_df = energy_df[energy_df['Rubrik'] == 'Endverbrauch - Total']

# Summarize energy data by year
total_energy_by_year = energy_df.groupby('Jahr')['TJ'].sum().reset_index()
total_energy_by_year.rename(columns={'TJ': 'TotalEnergy'}, inplace=True)

# Merge energy data with population data
final_merged_df = pd.merge(total_energy_by_year, person_df, on='Jahr')
final_merged_df['Energy_per_head'] = final_merged_df['TotalEnergy'] / final_merged_df['Bevölkerung']
final_merged_df['BIP_per_head'] = final_merged_df['BIP'] / final_merged_df['Bevölkerung']

# Calculate the annual change in BIP_per_head to find turning points
final_merged_df['BIP_change'] = final_merged_df['BIP_per_head'].pct_change()

# Identify turning points where the change in BIP_per_head changes sign
turning_points = final_merged_df[(final_merged_df['BIP_change'].shift(1) * final_merged_df['BIP_change'] < 0)]

# Create the plot
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot BIP per Head
color = 'tab:blue'
ax1.plot(final_merged_df['Jahr'], final_merged_df['BIP_per_head'], color=color, label='BIP_per_head')
ax1.set_xlabel('Jahr')
ax1.set_ylabel('BIP per Head', color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Mark turning points
ax1.plot(turning_points['Jahr'], turning_points['BIP_per_head'], 'go')  # 'go' for green circles

x_values = [final_merged_df['Jahr'].values[0], turning_points['Jahr'].values[0]]
y_values = [final_merged_df['BIP_per_head'].values[0], turning_points['BIP_per_head'].values[0]]
ax1.plot(x_values, y_values, 'g--')  # Green dashed line between points

# Draw lines between turning points
for i in range(len(turning_points['Jahr']) - 1):
    x_values = [turning_points['Jahr'].values[i], turning_points['Jahr'].values[i + 1]]
    y_values = [turning_points['BIP_per_head'].values[i], turning_points['BIP_per_head'].values[i + 1]]
    ax1.plot(x_values, y_values, 'g--')  # Green dashed line between points

# Plot Population on the primary y-axis
ax2 = ax1.twinx()
color = 'red'  # Set the color variable to red
ax2.plot(final_merged_df['Jahr'], final_merged_df['Bevölkerung'], color=color, label='Bevölkerung')  # Use 'color' variable
ax2.set_xlabel('Jahr')
ax2.set_ylabel('Bevölkerung', color=color)  # Set label color to red
ax2.tick_params(axis='y', labelcolor=color)  # Set tick color to red


# Title and legend
plt.title('Bevölkerung und BIP pro Kopf der Schweiz mit Wendepunkten (1980 - 2022)')
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

fig.tight_layout()
plt.show()
```
