### 3.1.2 Daten aufbereiten

Folgendes kann gemacht werden:

*   Unnötige Zeilen, Werte entfernen
*   Float zu Int umwandeln
*   Überschriften ändern

```python
# Erlaubte Werte für 'Rubrik' definieren
allowed_rubriken = [
    'Bruttoverbrauch',
    'Eigenverbrauch des Energiesektors, Netzverluste, Verbrauch der Speicherungen',
    'Endverbrauch - Dienstleistungen', 'Endverbrauch - Haushalte',
    'Endverbrauch - Industrie',
    'Endverbrauch - Statistische Differenz inkl. Landwirtschaft',
    'Endverbrauch - Total', 'Endverbrauch - Verkehr',
    'Energieumwandlung - Diverse Erneuerbare', 'Energieumwandlung - Gaswerke',
    'Energieumwandlung - Kernkraftwerke', 'Energieumwandlung - Raffinerien',
    'Energieumwandlung - Wasserkraftwerke',
    'Energieumwandlung - konventionell-thermische Kraft-, Fernheiz- und Fernheizkraftwerke',
    'Export', 'Import', 'Inlandproduktion', 'Lagerveränderung',
    'Nichtenergetischer Verbrauch'
]

# Erlaubte Werte für 'Energietraeger' definieren
allowed_energietraeger = [
    'Elektrizität', 'Erdölprodukte', 'Fernwärme', 'Gas', 'Holzenergie',
    'Kernbrennstoffe', 'Kohle', 'Müll und Industrieabfälle', 'Rohöl',
    'Wasserkraft', 'Uebrige erneuerbare Energien'
]

# Schema als Konstante definieren
BIP_SCHEMA = Schema([
    Column('Jahr', [CanConvertValidation(int), InRangeValidation(1980, 2024)]),
    Column('Bevölkerung', [CanConvertValidation(int), InRangeValidation(0, 1000000000)]),
    Column('BIP', [CanConvertValidation(float), InRangeValidation(0, float('inf'))]),
    Column('Bip_pro_Kopf', [CanConvertValidation(float), InRangeValidation(0, float('inf'))])
])

# Schema für das Energy DataFrame als Konstante definieren
ENERGY_SCHEMA = Schema([
    Column('Jahr', [CanConvertValidation(int), InRangeValidation(1980, 2024)]),  # Validiere, dass 'Jahr' ein Integer im spezifizierten Bereich ist
    Column('Rubrik', [CanConvertValidation(str), InListValidation(allowed_rubriken)]),  # Stelle sicher, dass 'Rubrik' ein String ist
    Column('Energietraeger', [CanConvertValidation(str), InListValidation(allowed_energietraeger)]),  # Stelle sicher, dass 'Energietraeger' ein String ist
    Column('TJ', [CanConvertValidation(float)])  # Stelle sicher, dass 'TJ' ein nicht-negativer Float ist
])

# Define the schema for CO2 emissions data as a constant
CO2_SCHEMA = Schema([
    Column('Jahr', [
        CanConvertValidation(int),  # Ensure 'Jahr' can be converted to an integer
        InRangeValidation(1980, 2024)  # Validate that 'Jahr' is within the 20th and 21st centuries
    ]),
    Column('THG', [
        CanConvertValidation(float)
    ])
])

def validate_and_clean_dataframe(df, schema):
    """
    Entfernt Spalten aus dem DataFrame, die nicht im Schema definiert sind.

    Parameters:
    - df (pd.DataFrame): Der zu validierende DataFrame.
    - schema (Schema): Das Schema, gegen das der DataFrame validiert wird.

    Returns:
    - df (pd.DataFrame): Der bereinigte DataFrame, der nur die im Schema definierten Spalten enthält.
    """
    # Extrahiere Spaltennamen aus dem Schema
    schema_columns = [column.name for column in schema.columns]

    # Entferne Spalten, die nicht im Schema definiert sind
    df = df.loc[:, schema_columns]
    return df

def validate_dataframe(df: pd.DataFrame, schema: Schema):
    """
    Validiert einen DataFrame gegen ein bereitgestelltes Schema und wirft eine Ausnahme, wenn Validierungsfehler gefunden werden.

    Parameters:
    - df (pd.DataFrame): Der zu validierende DataFrame.
    - schema (Schema): Das Schema, gegen das der DataFrame validiert wird.

    Raises:
    - ValueError: Wenn im DataFrame Validierungsfehler gefunden werden.

    Returns:
    - None
    """
    # Validiere den DataFrame mit dem bereitgestellten Schema
    errors = schema.validate(df)
    if errors:
        # Sammle alle Fehlermeldungen in einem einzigen String
        error_messages = "\n".join(str(e) for e in errors)
        # Drucke Fehler zur Protokollierung
        print(f"Validation errors occurred:\n{error_messages}")
        # Wirf eine Ausnahme mit allen Fehlermeldungen
        raise ValueError(f"Validation errors occurred:\n{error_messages}")
    else:
        print("Alle Daten sind gültig.")
```

### 3.1.3 Daten organisieren

Daten zu sinnvollen Datasets zusammenfügen?

*   BevölkerungsEntwicklung
*   BIPEntwicklung
*   SektorenEntwicklung
*   C02Ausstoss




```python
def summarize_energy_data_per_year_per_column(df, category_column):
    """
    Summiert Energieverbrauchsdaten nach Jahr und einer spezifizierten Kategorie und berechnet die Prozentsätze jeder Kategorie.

    Parameters:
    df (pd.DataFrame): Der DataFrame, der die Energieverbrauchsdaten enthält.
    category_column (str): Der Name der Spalte, die die Kategorien beschreibt, nach denen summiert wird.

    Returns:
    pd.DataFrame: Ein DataFrame mit den summierten Daten und den berechneten Prozentsätzen für jede Kategorie.
    """
    # Gruppiere Daten nach Jahr und der spezifizierten Kategorie, und summiere die Energieverbrauchsdaten ('TJ') für jede Gruppe
    category_data_summed_per_year = df.groupby(['Jahr', category_column])['TJ'].sum().reset_index()

    # Gruppiere Daten nur nach Jahr und summiere die gesamte Energie ('TJ') für jedes Jahr
    total_energy_per_year = df.groupby('Jahr')['TJ'].sum().reset_index()
    total_energy_per_year.rename(columns={'TJ': 'TotalEnergy'}, inplace=True)  # Benenne die Spalte 'TJ' um in 'TotalEnergy'

    # Führe die Daten zu summierten Kategorien mit den totalen Energie-Daten zusammen, basierend auf dem Jahr
    summarized_data = pd.merge(category_data_summed_per_year, total_energy_per_year, on='Jahr')

    # Berechne den Prozentsatz des Verbrauchs jeder Kategorie relativ zur gesamten Energieverbrauchsmenge des jeweiligen Jahres
    summarized_data['Percentage'] = (summarized_data['TJ'] / summarized_data['TotalEnergy']) * 100

    return summarized_data
```

```python
EnergiedatenEndverbrauch = Energiedaten[Energiedaten['Rubrik'] == 'Endverbrauch - Total']

EnergieDatenEndverbrauchSummiert = summarize_energy_data_per_year_per_column(EnergiedatenEndverbrauch, 'Energietraeger')

print(EnergieDatenEndverbrauchSummiert)
```

```python
# Apply correct filtering for specific categories
EnergiedatenSektoren = Energiedaten[(Energiedaten['Rubrik'] == 'Endverbrauch - Haushalte') |
        (Energiedaten['Rubrik'] == 'Endverbrauch - Dienstleistungen') |
        (Energiedaten['Rubrik'] == 'Endverbrauch - Industrie') |
        (Energiedaten['Rubrik'] == 'Endverbrauch - Statistische Differenz inkl. Landwirtschaft') |
        (Energiedaten['Rubrik'] == 'Endverbrauch - Verkehr')]

EnergieDatenSektorenSummiert = summarize_energy_data_per_year_per_column(EnergiedatenSektoren, 'Rubrik')

print(EnergieDatenSektorenSummiert)
```

```python
# Load data
df = pd.read_csv('https://raw.githubusercontent.com/michaelzimmermannstudhslu/BINA_2024_energy_research/main/data/energiebilanz_ch.csv')

# Clean data
df.dropna(inplace=True)
df['TJ'] = pd.to_numeric(df['TJ'], errors='coerce')
df['TJ'] = df['TJ'].abs()
df.dropna(subset=['TJ'], inplace=True)
df['TJ'] = df['TJ'].astype(int)

# Apply correct filtering
df = df[((df['Rubrik'] == 'Inlandproduktion') | (df['Rubrik'] == 'Import') ) &
        (df['Energietraeger'] != 'Wasserkraft') &
        (df['Energietraeger'] != 'Uebrige erneuerbare Energien')]


# Group by year and Energieträger, sum up TJ
energy_sum_by_year_and_type = df.groupby(['Jahr', 'Energietraeger'])['TJ'].sum().reset_index()

# Group by year only, sum up TJ for total energy
total_energy_by_year = df.groupby('Jahr')['TJ'].sum().reset_index()
total_energy_by_year.rename(columns={'TJ': 'TotalEnergy'}, inplace=True)

# Merge while calculating the percentage
merged_df = pd.merge(energy_sum_by_year_and_type, total_energy_by_year, on='Jahr')
merged_df['Percentage'] = (merged_df['TJ'] / merged_df['TotalEnergy']) * 100

# Pivot the data for plotting by Energieträger
pivoted_data = merged_df.pivot_table(index='Jahr', columns='Energietraeger', values='Percentage', fill_value=0)

# Create a stacked area plot
plt.figure(figsize=(12, 8))
colors = ['#ffd700', '#ffb14e', '#fa8775', '#ea5f94', '#cd34b5', '#9d02d7', '#0000ff', '#0000a2', '#000000']
plt.stackplot(pivoted_data.index, pivoted_data.T, labels=pivoted_data.columns, colors=colors)
plt.title('Gestapeltes Flächendiagramm Energieproduktion Energieträger')
plt.xlabel('Year')
plt.ylabel('Gesamtproduktion in Prozent')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title='Energieträger')
plt.show()
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import itertools

# Load the energy data from an online source into a table called a DataFrame.
df = pd.read_csv('https://raw.githubusercontent.com/michaelzimmermannstudhslu/BINA_2024_energy_research/main/data/energiebilanz_ch.csv')

# Remove any rows that have missing data to ensure data quality.
df = df.dropna()

# Convert energy usage data to a numeric type, forcing errors to be recognized as NaN (not a number), and make all values positive.
df['TJ'] = pd.to_numeric(df['TJ'], errors='coerce')
df['TJ'] = df['TJ'].abs()

# Remove any rows where energy usage data couldn't be converted to numbers.
df = df.dropna(subset=['TJ'])

# Only consider data where the category is 'Endverbrauch - Total', which stands for 'Total final consumption'.
df = df[df['Rubrik'] == 'Endverbrauch - Total']

# Group data by year and type of energy, and sum up all energy usage values.
energy_sum_by_year_and_type = df.groupby(['Jahr', 'Energietraeger'])['TJ'].sum().reset_index()
total_energy_by_year = df.groupby('Jahr')['TJ'].sum().reset_index()

# Rename the column for clearer understanding.
total_energy_by_year.rename(columns={'TJ': 'TotalEnergy'}, inplace=True)

# Merge the two datasets on the year column, so each row shows total energy and energy by type for each year.
merged_df = pd.merge(energy_sum_by_year_and_type, total_energy_by_year, on='Jahr')

# Calculate what percentage of the total energy each type represents per year.
merged_df['Percentage'] = (merged_df['TJ'] / merged_df['TotalEnergy']) * 100

# Prepare data for the model by creating dummy variables for each type of energy (one-hot encoding).
df_for_model = merged_df[['Jahr', 'Energietraeger', 'Percentage']].copy()
df_for_model = pd.get_dummies(df_for_model, columns=['Energietraeger'])

# Separate the data into input features (X) and output targets (y), where y is the percentage to predict.
X = df_for_model.drop('Percentage', axis=1).values
y = df_for_model['Percentage'].values

# Standardize features by removing the mean and scaling to unit variance to help the model learn more effectively.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training set and test set, where 20% of the data is reserved for testing the model.
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Convert arrays to tensors, which are a type of data structure used in PyTorch for building models.
X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32)

# Define a simple linear regression model using a neural network framework.
class LinearRegressionModel(nn.Module):
    def __init__(self, input_size):
        super(LinearRegressionModel, self).__init__()
        self.linear = nn.Linear(input_size, 1)

    def forward(self, x):
        return self.linear(x)

# Initialize the model with the correct number of input features.
model = LinearRegressionModel(X_train.shape[1])

# Define a mean squared error loss function and the stochastic gradient descent optimizer.
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Train the model for 100 epochs, updating weights to minimize the loss.
epochs = 100
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()

    # Perform a forward pass (calculate predicted y from x), and compute the loss.
    outputs = model(X_train).squeeze()
    loss = criterion(outputs, y_train)

    # Perform a backward pass to compute gradients, and update model weights.
    loss.backward()
    optimizer.step()

    # Print the loss every 10 epochs to monitor progress.
    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

# Prepare data for predicting future energy percentages for years 2025-2051 for each type of energy.
unique_energy_types = merged_df['Energietraeger'].unique()
future_years = np.arange(2025, 2051)
future_data = pd.DataFrame(list(itertools.product(future_years, unique_energy_types)), columns=['Jahr', 'Energietraeger'])
future_data = pd.get_dummies(future_data, columns=['Energietraeger'])

# Add missing columns with zeros in case some energy types were not present in the future data.
missing_cols = set(df_for_model.columns) - set(future_data.columns) - {'Percentage'}
for col in missing_cols:
    future_data[col] = 0

# Scale the future data features and convert to tensor.
future_features = scaler.transform(future_data)
future_features = torch.tensor(future_features, dtype=torch.float32)

# Use the trained model to predict future energy percentages.
model.eval()
with torch.no_grad():
    future_predictions = model(future_features).squeeze()

# Clamp output to ensure no negative percentages.
outputs = torch.clamp(outputs, min=0, max=100)

# Store the predicted percentages back in the data frame.
future_data['Predicted_Percentage'] = future_predictions.numpy()

# Function to revert one-hot encoded columns back to a single categorical column for easier interpretation.
def revert_dummies(revert_df, prefix):
    category_cols = [col for col in revert_df if col.startswith(prefix)]
    revert_df['Energietraeger'] = revert_df[category_cols].idxmax(axis=1)
    revert_df['Energietraeger'] = revert_df['Energietraeger'].apply(lambda x: x.replace(prefix + '_', ''))
    revert_df.drop(columns=category_cols, inplace=True)
    return revert_df

# Apply the function to revert dummy variables.
real_future_data = revert_dummies(future_data, 'Energietraeger')
real_future_data['Predicted_Percentage'] = real_future_data['Predicted_Percentage'].clip(lower=0)
print(real_future_data)

# Function to plot a histogram showing the distribution of predicted energy usage percentages by energy type for a selected year.
def plot_histogram(year):
    filtered_data = real_future_data[real_future_data['Jahr'] == year]
    colors = plt.cm.viridis(np.linspace(0, 1, len(filtered_data['Energietraeger'].unique())))
    plt.figure(figsize=(10, 6))
    bars = plt.bar(filtered_data['Energietraeger'], filtered_data['Predicted_Percentage'], color=colors)
    plt.xlabel('Energy Type')
    plt.ylabel('Energy Used (TJ)')
    plt.title(f'Energy Usage Distribution for {year}')
    plt.xticks(rotation=45)
    plt.ylim(0, 100)
    for bar, label in zip(bars, filtered_data['Predicted_Percentage']):
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{label:.2f}%', ha='center', va='bottom', color='black', fontsize=9)
    plt.show()

# Create a dropdown menu for selecting a year to visualize, and link it to the plotting function.
year_dropdown = widgets.Dropdown(
    options=real_future_data['Jahr'].unique(),
    value=real_future_data['Jahr'].unique()[0],
    description='Year:',
    disabled=False,
)

widgets.interactive(plot_histogram, year=year_dropdown)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
person_df = pd.read_csv('https://raw.githubusercontent.com/michaelzimmermannstudhslu/BINA_2024_energy_research/main/data/bip_total.csv')
energy_df = pd.read_csv('https://raw.githubusercontent.com/michaelzimmermannstudhslu/BINA_2024_energy_research/main/data/energiebilanz_ch.csv')

# Clean and process energy data
energy_df.dropna(inplace=True)
energy_df['TJ'] = pd.to_numeric(energy_df['TJ'], errors='coerce')
energy_df['TJ'] = energy_df['TJ'].abs()
energy_df.dropna(subset=['TJ'], inplace=True)
energy_df['TJ'] = energy_df['TJ'].astype(int)
energy_df = energy_df[energy_df['Rubrik'] == 'Endverbrauch - Total']

# Summarize energy data by year
total_energy_by_year = energy_df.groupby('Jahr')['TJ'].sum().reset_index()
total_energy_by_year.rename(columns={'TJ': 'TotalEnergy'}, inplace=True)

# Merge energy data with population data
final_merged_df = pd.merge(total_energy_by_year, person_df, on='Jahr')
final_merged_df['Energy_per_head'] = final_merged_df['TotalEnergy'] / final_merged_df['Bevölkerung']
final_merged_df['BIP_per_head'] = final_merged_df['BIP'] / final_merged_df['Bevölkerung']

# Calculate the annual change in BIP_per_head to find turning points
final_merged_df['BIP_change'] = final_merged_df['BIP_per_head'].pct_change()

# Identify turning points where the change in BIP_per_head changes sign
turning_points = final_merged_df[(final_merged_df['BIP_change'].shift(1) * final_merged_df['BIP_change'] < 0)]

# Create the plot
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot BIP per Head
color = 'tab:blue'
ax1.plot(final_merged_df['Jahr'], final_merged_df['BIP_per_head'], color=color, label='BIP_per_head')
ax1.set_xlabel('Jahr')
ax1.set_ylabel('BIP per Head', color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Mark turning points
ax1.plot(turning_points['Jahr'], turning_points['BIP_per_head'], 'go')  # 'go' for green circles

x_values = [final_merged_df['Jahr'].values[0], turning_points['Jahr'].values[0]]
y_values = [final_merged_df['BIP_per_head'].values[0], turning_points['BIP_per_head'].values[0]]
ax1.plot(x_values, y_values, 'g--')  # Green dashed line between points

# Draw lines between turning points
for i in range(len(turning_points['Jahr']) - 1):
    x_values = [turning_points['Jahr'].values[i], turning_points['Jahr'].values[i + 1]]
    y_values = [turning_points['BIP_per_head'].values[i], turning_points['BIP_per_head'].values[i + 1]]
    ax1.plot(x_values, y_values, 'g--')  # Green dashed line between points

# Plot Population on the primary y-axis
ax2 = ax1.twinx()
color = 'red'  # Set the color variable to red
ax2.plot(final_merged_df['Jahr'], final_merged_df['Bevölkerung'], color=color, label='Bevölkerung')  # Use 'color' variable
ax2.set_xlabel('Jahr')
ax2.set_ylabel('Bevölkerung', color=color)  # Set label color to red
ax2.tick_params(axis='y', labelcolor=color)  # Set tick color to red


# Title and legend
plt.title('Bevölkerung und BIP pro Kopf der Schweiz mit Wendepunkten (1980 - 2022)')
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/michaelzimmermannstudhslu/BINA_2024_energy_research/main/data/bip_total.csv')
df = df.dropna()

# Initialize the linear regression model
model_population = LinearRegression()
model_gdp = LinearRegression()

# Prepare data for regression (reshape year data for fitting)
X = df['Jahr'].values.reshape(-1, 1)  # Features
y_population = df['Bevölkerung'].values  # Target for population
y_gdp = df['BIP'].values  # Target for GDP

# Fit the models
model_population.fit(X, y_population)
model_gdp.fit(X, y_gdp)

# Predict the values
y_population_pred = model_population.predict(X)
y_gdp_pred = model_gdp.predict(X)

fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot population data
color = 'tab:red'
ax1.set_xlabel('Jahr')
ax1.set_ylabel('Bevölkerung', color=color)
ax1.plot(df['Jahr'], df['Bevölkerung'], marker='o', color=color, label='Actual Bevölkerung')
ax1.plot(df['Jahr'], y_population_pred, color=color, linestyle='--', label='Regression Line - Bevölkerung')
ax1.tick_params(axis='y', labelcolor=color)

# Second y-axis for GDP
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('BIP', color=color)
ax2.plot(df['Jahr'], df['BIP'], marker='o', color=color, label='Actual BIP')
ax2.plot(df['Jahr'], y_gdp_pred, color=color, linestyle='--', label='Regression Line - BIP')
ax2.tick_params(axis='y', labelcolor=color)

# Title and legend
plt.title('Bevölkerung und BIP der Schweiz mit Regression (1980 - 2022)')
fig.tight_layout()

# Handling legends for both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left')
ax2.legend(lines + lines2, labels + labels2, loc='upper left')
plt.show()
```
