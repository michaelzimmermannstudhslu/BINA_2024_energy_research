### 3.1.2 Daten aufbereiten

In diesem Abschnitt werden die Daten für die Datenorganisation vorbereitet, damit sie besser organisiert werden können.

```python
# Die Daten des Endverbrauchs sind nicht ganz richtig. Der Endverbrauch - Total in Bezug der Energieträger umfasst: Elektrizität, Erdölprodukte, Fernwärme, Gas, Holzenergie, Kohle, Müll und Industrieabfälle und Übrige erneuerbare Energien.
# Hier müssen Kernbrennstoffe, Rohöl und Wasserkraft entfernt werden

# Erlaubte Werte für 'Rubrik' definieren
allowed_rubriken = [
    'Bruttoverbrauch',
    'Eigenverbrauch des Energiesektors, Netzverluste, Verbrauch der Speicherungen',
    'Endverbrauch - Total',
    'Export', 'Import', 'Inlandproduktion', 'Lagerveränderung',
    'Nichtenergetischer Verbrauch'
]

enverbrauch_rubriken = [
    'Endverbrauch - Dienstleistungen', 'Endverbrauch - Haushalte',
    'Endverbrauch - Industrie',
    'Endverbrauch - Statistische Differenz inkl. Landwirtschaft',
    'Endverbrauch - Verkehr'
]

# Erlaubte Werte für 'Energieträger' definieren, aufgeteilt in erneuerbare und nicht-erneuerbare Energiequellen
renewable_rubriken = [
    'Energieumwandlung - Wasserkraftwerke', 'Energieumwandlung - Diverse Erneuerbare'
]

non_renewable_rubriken = [
    'Energieumwandlung - Gaswerke', 'Energieumwandlung - Kernkraftwerke', 'Energieumwandlung - Raffinerien', 'Energieumwandlung - konventionell-thermische Kraft-, Fernheiz- und Fernheizkraftwerke',
]

# Erlaubte Werte für 'Energieträger' definieren, aufgeteilt in erneuerbare und nicht-erneuerbare Energiequellen
renewable_energietraeger = [
    'Wasserkraft', 'Uebrige erneuerbare Energien', 'Holzenergie'
]

non_renewable_energietraeger = [
    'Elektrizität', 'Erdölprodukte', 'Fernwärme', 'Gas',
    'Kernbrennstoffe', 'Kohle', 'Müll und Industrieabfälle', 'Rohöl'
]

# Definition des Schemas für Daten über das Bruttoinlandsprodukt und Bevölkerung
BIP_SCHEMA = Schema([
    Column('Jahr', [CanConvertValidation(int), InRangeValidation(1980, 2024)]),
    Column('Bevölkerung', [CanConvertValidation(int), InRangeValidation(0, 1000000000)]),
    Column('BIP', [CanConvertValidation(float), InRangeValidation(0, float('inf'))]),
    Column('Bip_pro_Kopf', [CanConvertValidation(float), InRangeValidation(0, float('inf'))])
])

# Definition des Schemas für das Energie-Daten-DataFrame mit Überprüfung auf die erlaubten Rubriken und Energieträger
ENERGY_SCHEMA = Schema([
    Column('Jahr', [CanConvertValidation(int), InRangeValidation(1980, 2024)]),
    Column('Rubrik', [CanConvertValidation(str), InListValidation(allowed_rubriken + enverbrauch_rubriken + renewable_rubriken + non_renewable_rubriken)]),
    Column('Energietraeger', [CanConvertValidation(str), InListValidation(renewable_energietraeger + non_renewable_energietraeger)]),
    Column('TJ', [CanConvertValidation(float)])
])

# Definition des Schemas für CO2-Emissionsdaten
CO2_SCHEMA = Schema([
    Column('Jahr', [CanConvertValidation(int), InRangeValidation(1980, 2024)]),
    Column('THG', [CanConvertValidation(float)])
])

# Definition des Schemas für CO2-Emissionsdaten über verschiedene Sektoren hinweg
CO2_SECTOR_SCHEMA = Schema([
    Column('Jahr', [CanConvertValidation(int), InRangeValidation(1980, 2024)]),
    Column('Gebäude', [CanConvertValidation(float), InRangeValidation(0, float('inf'))]),
    Column('Verkehr', [CanConvertValidation(float), InRangeValidation(0, float('inf'))]),
    Column('Industrie', [CanConvertValidation(float), InRangeValidation(0, float('inf'))]),
    Column('Übrige', [CanConvertValidation(float), InRangeValidation(0, float('inf'))])
])

def validate_and_clean_dataframe(df, schema):
    """
    Überprüft und bereinigt einen DataFrame basierend auf einem gegebenen Schema. Entfernt Spalten, die nicht im Schema definiert sind.

    Parameter:
    - df (pd.DataFrame): Der zu validierende DataFrame.
    - schema (Schema): Das Schema, gegen das der DataFrame validiert wird.

    Rückgabe:
    - df (pd.DataFrame): Der bereinigte DataFrame, der nur die im Schema definierten Spalten enthält.
    """
    schema_columns = [column.name for column in schema.columns]  # Extrahiere Spaltennamen aus dem Schema
    df = df.loc[:, schema_columns]  # Behalte nur die Spalten, die im Schema definiert sind
    return df

def validate_dataframe(df: pd.DataFrame, schema: Schema):
    """
    Validiert einen DataFrame gegen ein bereitgestelltes Schema. Wirft eine Ausnahme, wenn Validierungsfehler gefunden werden.

    Parameter:
    - df (pd.DataFrame): Der zu validierende DataFrame.
    - schema (Schema): Das Schema, gegen das der DataFrame validiert wird.

    Fehler:
    - ValueError: Wird ausgelöst, wenn im DataFrame Validierungsfehler gefunden werden.

    Rückgabe:
    - None
    """
    errors = schema.validate(df)  # Validiere den DataFrame mit dem bereitgestellten Schema
    if errors:
        error_messages = "\n".join(str(e) for e in errors)  # Sammle alle Fehlermeldungen in einem String
        print(f"Validation errors occurred:\n{error_messages}")  # Drucke Fehler zur Protokollierung
        raise ValueError(f"Validation errors occurred:\n{error_messages}")  # Wirf eine Ausnahme mit allen Fehlermeldungen
    else:
        print("Alle Daten sind gültig.")  # Bestätigung, dass alle Daten gültig sind, wenn keine Fehler gefunden wurden
```

```python
# Vorverarbeitung und Konvertierung von Datentypen
# Konvertierung der 'Jahr' Spalte zu numerischen Werten, wobei Fehler ignoriert und durch 0 ersetzt werden
Energiedaten['Jahr'] = pd.to_numeric(Energiedaten['Jahr'], errors='coerce').fillna(0).astype(int)

# Konvertierung der 'TJ' Spalte zu numerischen Werten, Fehler werden ebenfalls ignoriert und durch 0 ersetzt
Energiedaten['TJ'] = pd.to_numeric(Energiedaten['TJ'], errors='coerce').fillna(0).astype(int)

# Ersetzen von Zeichensetzung in der 'Rubrik' Spalte, um Konsistenz zu gewährleisten
# Beispielsweise Ersetzen von '–' durch '-', um Einheitlichkeit in der Darstellung sicherzustellen
Energiedaten['Rubrik'] = Energiedaten['Rubrik'].str.replace('–', '-', regex=False)

# Entfernen aller Zeilen mit fehlenden Daten (NaN-Werten), um die Datenqualität für die Analyse zu erhöhen
Energiedaten.dropna(inplace=True)

# Anwenden der definierten Validierungs- und Bereinigungsfunktion, um das Daten-DataFrame an das festgelegte Schema anzupassen
Energiedaten = validate_and_clean_dataframe(Energiedaten, ENERGY_SCHEMA)

# Validierung des bereinigten DataFrames gegen das festgelegte Schema
# Überprüfung auf Einhaltung der Datentypen und Wertebereiche laut Schema
validate_dataframe(Energiedaten, ENERGY_SCHEMA)

# Ausgabe des finalen, validierten DataFrames
# Zeigt die strukturierten und bereinigten Daten gemäß dem Schema
print(Energiedaten)
```

In diesem Abschnitt werden die vorbereiteten Daten organisiert. Dabei werden sinnvolle Datasets zusammengestellt, damit die entsprechenden Daten in notwendigen Qualität und Quantität für die Analyse und die Präsentation zur Verfügung stehen und effizient verwendet werden können.


```python
def summarize_energy_data_per_year_per_column(df, category_column):
    """
    Summiert die Energieverbrauchsdaten pro Jahr und Kategorie und berechnet die prozentuale Verteilung jeder Kategorie im Vergleich zum Gesamtverbrauch pro Jahr.

    Parameters:
    df (pd.DataFrame): Der DataFrame, der die Energieverbrauchsdaten enthält.
    category_column (str): Der Name der Spalte, die die Kategorien beschreibt, nach denen summiert wird.

    Returns:
    pd.DataFrame: Ein DataFrame, der die summierten Daten und die berechneten Prozentsätze für jede Kategorie enthält.
    """
    # Sicherstellen, dass die 'TJ' Spalte numerische Werte enthält
    df['TJ'] = pd.to_numeric(df['TJ'], errors='coerce')
    df.dropna(subset=['TJ'], inplace=True)  # Entfernen von Zeilen, wo 'TJ' NaN ist nach der Konversion

    # Gruppieren der Daten nach Jahr und der angegebenen Kategorie und Summierung der Energieverbrauchswerte ('TJ') für jede Gruppe
    category_data_summed_per_year = df.groupby(['Jahr', category_column])['TJ'].sum().reset_index()

    # Gruppieren der Daten nur nach Jahr und Summierung der gesamten Energie ('TJ') für jedes Jahr
    total_energy_per_year = df.groupby('Jahr')['TJ'].sum().reset_index()
    total_energy_per_year.rename(columns={'TJ': 'TotalEnergy'}, inplace=True)  # Umbenennen der Spalte 'TJ' in 'TotalEnergy' für Klarheit

    # Zusammenführen der summierten Kategoriedaten mit den totalen Energie-Daten anhand des Jahres
    summarized_data = pd.merge(category_data_summed_per_year, total_energy_per_year, on='Jahr')

    # Berechnung des prozentualen Anteils des Verbrauchs jeder Kategorie im Vergleich zum Gesamtenergieverbrauch des jeweiligen Jahres
    summarized_data['Percentage'] = (summarized_data['TJ'] / summarized_data['TotalEnergy']) * 100

    # Rückgabe des bearbeiteten DataFrames, der nun sowohl die absoluten Werte als auch die Prozentsätze enthält
    return summarized_data
```

```python
def plot_pie(df, year, carrier_col='Energietraeger', percentage_col='Percentage', plot_title=None, legend_title='Energieträger', colors=None):
    """
    Erstellt ein Tortendiagramm des Energieverbrauchs nach bestimmten Energieträgern für ein vorgegebenes Jahr aus einem bereitgestellten DataFrame.

    Parameter:
    - df (DataFrame): Der DataFrame, der die Energiedaten enthält.
    - year (int): Das Jahr, für das die Daten visualisiert werden sollen.
    - carrier_col (str): Spaltenname im DataFrame, der die Energieträger kennzeichnet.
    - percentage_col (str): Spaltenname im DataFrame, der die Prozentsätze darstellt.
    - plot_title (str): Titel des Diagramms, falls None, wird ein Standardtitel generiert.
    - legend_title (str): Titel für die Legende.
    - colors (list): Liste von Farben für die Segmente des Tortendiagramms; falls None, wird eine Standardfarbpalette verwendet.

    Der DataFrame sollte mindestens die Spalten nach `carrier_col` und `percentage_col` neben 'Jahr' enthalten.
    """
    # Daten für das ausgewählte Jahr filtern
    filtered_data = df[df['Jahr'] == year]
    labels = filtered_data[carrier_col]
    sizes = filtered_data[percentage_col]

    # Standardtitel setzen, falls keiner angegeben wurde
    if plot_title is None:
        plot_title = f'Energieverbrauch nach {legend_title} {year}'

    # Farben durch eine Farbpalette zuweisen, falls keine Farben angegeben wurden
    if colors is None:
        cmap = plt.get_cmap('tab20')  # Verbesserte Farbpalette
        colors = [cmap(i) for i in np.linspace(0, 1, len(labels))]

    # Das größte Segment leicht hervorheben
    explode = [0.1] * len(sizes)

    # Tortendiagramm erstellen
    plt.figure(figsize=(10, 8))
    patches, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode, pctdistance=1.1)

    # Beschriftungen und Prozentangaben verbessern
    for text, autotext in zip(texts, autotexts):
        text.set_fontsize(10)
        text.set_color('white')  # Farbe der Texte zu Grau ändern, um Lesbarkeit zu verbessern
        autotext.set_fontsize(10)

    # Legende hinzufügen
    plt.legend(patches, labels, title=legend_title, loc="best", bbox_to_anchor=(1, 0, 0.5, 1))
    plt.title(plot_title)
    plt.axis('equal')  # Gleiches Seitenverhältnis stellt sicher, dass das Pie als Kreis gezeichnet wird.
    plt.show()


def plot_stacked_area(df, x_label, y_label, title, colors, legend_title):
    """
    Erstellt ein gestapeltes Flächendiagramm.

    Parameter:
    - df (DataFrame): DataFrame, der die Daten für das Diagramm enthält, indiziert durch die Werte der x-Achse.
    - x_label (str): Beschriftung der x-Achse.
    - y_label (str): Beschriftung der y-Achse.
    - title (str): Titel des Diagramms.
    - colors (list): Liste von Farben für die verschiedenen Stapel.
    - legend_title (str): Titel für die Legende.
    """
    plt.figure(figsize=(12, 8))
    plt.stackplot(df.index, df.T, labels=df.columns, colors=colors)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title=legend_title)
    plt.show()

def plot_stacked_bar(data, title, xlabel, ylabel, legend_title='Energy Types', figsize=(12, 8), colors=None):
    """
    Erstellt ein gestapeltes Balkendiagramm mit gegebenen Daten.

    Parameter:
    - data (DataFrame): Ein pandas DataFrame, wobei jede Spalte eine unterschiedliche Kategorie darstellt und der Index die x-Achse repräsentiert.
    - title (str): Der Titel des Diagramms.
    - xlabel (str): Beschriftung der x-Achse.
    - ylabel (str): Beschriftung der y-Achse.
    - legend_title (str): Titel für die Legende.
    - figsize (tuple): Tupel für die Größe der Figur (Breite, Höhe).
    - colors (list): Eine Liste von Farben, die für die verschiedenen Stapel verwendet werden sollen.
                     Wenn 'None' übergeben wird, werden automatisch Farben aus der Farbpalette verwendet.
    """
    fig, ax = plt.subplots(figsize=figsize)
    bottom = np.zeros(len(data))  # Dies wird die kumulative Untergrenze für jede Stapelung verfolgen

    if colors is None:
        # Standardfarbpalette verwenden, wenn keine Farben angegeben wurden
        cmap = plt.get_cmap('tab20')
        colors = [cmap(i) for i in np.linspace(0, 1, len(data.columns))]

    # Jede Spalte als separater Teil des Stapels plotten, Farben zuteilen
    for idx, col in enumerate(data.columns):
        ax.bar(data.index, data[col], bottom=bottom, label=col, color=colors[idx % len(colors)])
        bottom += data[col].values  # Aktualisieren der Untergrenze für den nächsten Stapel

    # Titel und Beschriftungen einstellen
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(title=legend_title, loc='upper left', bbox_to_anchor=(1, 1))

    # Diagramm anzeigen
    plt.show()

def plot_multiple_lines(data, title, xlabel, ylabel, legend_title='Energy Types', marker='o', figsize=(12, 8)):
    """
    Zeichnet mehrere Linien auf einem einzigen Diagramm.

    Parameter:
    - data (DataFrame): Ein pandas DataFrame, wobei jede Spalte eine unterschiedliche Linie darstellt und der Index die x-Achse repräsentiert.
    - title (str): Der Titel des Diagramms.
    - xlabel (str): Beschriftung der x-Achse.
    - ylabel (str): Beschriftung der y-Achse.
    - legend_title (str): Titel für die Legende.
    - marker (str): Markierungsstil für die Datenpunkte auf den Linien.
    - figsize (tuple): Tupel für die Größe der Figur (Breite, Höhe).
    """
    fig, ax = plt.subplots(figsize=figsize)

    # Jede Spalte als separate Linie plotten
    for col in data.columns:
        ax.plot(data.index, data[col], marker=marker, label=col)

    # Diagramm mit Titeln und Beschriftungen einrichten
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(title=legend_title, loc='upper left', bbox_to_anchor=(1, 1))

    # y-Achse beginnt bei 0
    ax.set_ylim(bottom=0)

    # Diagramm anzeigen
    plt.show()
```

Um die erste Fragestellung in Bezug der Energieträgerentwicklung zu beantworten, werden die Parameter des Endverbrauchs nach Energieträger analysiert.

```python
# Beschreibung der Verteilung der gesamten Energiemenge ('TotalEnergy')
# Diese Statistik gibt einen Überblick über wichtige Maßzahlen wie Durchschnitt, Standardabweichung, Minima und Maxima.
print(EnergieDatenEndverbrauchSummiert['TotalEnergy'].describe())

# Beschreibung der Verteilung der Energiemengen ('TJ') für alle Einträge
# Hier werden ähnliche Maßzahlen für die Spalte 'TJ' berechnet, die den Energieverbrauch in Terajoule darstellt.
print(EnergieDatenEndverbrauchSummiert['TJ'].describe())

# Ermittlung des Jahres mit dem maximalen Gesamtenergieverbrauch
# Das Jahr mit dem höchsten Verbrauch sowie der entsprechende Energieverbrauchswert werden identifiziert.
max_energy_year = EnergieDatenEndverbrauchSummiert.loc[EnergieDatenEndverbrauchSummiert['TotalEnergy'].idxmax(), 'Jahr']
max_energy_value = EnergieDatenEndverbrauchSummiert['TotalEnergy'].max()

# Ermittlung des Jahres mit dem minimalen Gesamtenergieverbrauch
# Analog wird das Jahr mit dem geringsten Verbrauch sowie der entsprechende Wert ermittelt.
min_energy_year = EnergieDatenEndverbrauchSummiert.loc[EnergieDatenEndverbrauchSummiert['TotalEnergy'].idxmin(), 'Jahr']
min_energy_value = EnergieDatenEndverbrauchSummiert['TotalEnergy'].min()

print(f"Das Jahr mit dem höchsten Gesamtenergieverbrauch ist {max_energy_year} mit {max_energy_value} TJ.")
print(f"Das Jahr mit dem geringsten Gesamtenergieverbrauch ist {min_energy_year} mit {min_energy_value} TJ.")

# Pivotieren der Daten, um jährliche Summen für jeden Energieträger zu erhalten
# Diese Pivot-Tabelle ermöglicht einen detaillierten Überblick über den Energieverbrauch pro Jahr und Energieträger.
energie_pivot_data = EnergieDatenEndverbrauchSummiert.pivot_table(index='Jahr', columns='Energietraeger', values='TJ', aggfunc='sum')

# Berechnung der Veränderung des Energieverbrauchs zwischen dem ersten und letzten Jahr für jeden Energieträger
# Diese Berechnung hilft, die Trends im Energieverbrauch über den gesamten Zeitraum zu verstehen.
change_in_energy = energie_pivot_data.iloc[-1] - energie_pivot_data.iloc[0]

# Identifikation des Energieträgers mit der größten Zunahme
# Hier wird der Energieträger mit der signifikantesten Erhöhung im Verbrauch und der zugehörige Wert festgestellt.
max_change_rubrik = change_in_energy.idxmax()
max_change_value = change_in_energy.max()

# Identifikation des Energieträgers mit der größten Abnahme
# Ebenso wird der Energieträger mit dem größten Rückgang im Verbrauch und der zugehörige Wert bestimmt.
min_change_rubrik = change_in_energy.idxmin()
min_change_value = change_in_energy.min()

print(f"Der Energieträger mit der größten Zunahme im Energieverbrauch ist {max_change_rubrik} mit einer Veränderung von {max_change_value} TJ.")
print(f"Der Energieträger mit dem größten Rückgang im Energieverbrauch ist {min_change_rubrik} mit einer Veränderung von {min_change_value} TJ.")
```

```python
# Plotting the area chart
fig, ax = plt.subplots(figsize=(10, 6))
CO2Ausstoss_Sektoren.plot(kind='area', stacked=True, ax=ax)

# Calculate cumulative sum for total contributions per year
total_contributions = CO2Ausstoss_Sektoren.sum(axis=1)

# Use numpy to calculate a smoothed version of the total contributions
smoothed = savgol_filter(total_contributions, 9, 3)  # Using Savitzky-Golay filter for better smoothing

# Calculate second derivative
second_derivative = np.diff(smoothed, n=2)

# Define a threshold for "big changes"
threshold = np.percentile(np.abs(second_derivative), 95)  # top 5% magnitude changes

# Identify inflection points (zero crossings in the second derivative)
inflection_points = np.where(np.diff(np.sign(second_derivative)))[0] + 1  # Adjust for shift caused by diff

# Plot inflection points
for point in inflection_points:
    if point < len(CO2Ausstoss_Sektoren) and np.abs(second_derivative[point - 1]) > threshold:
        year = CO2Ausstoss_Sektoren.index[point]
        ax.plot(year, total_contributions.loc[year], 'ro')  # Mark inflection point with a red circle
        ax.axvline(x=year, color='black', linestyle='--', linewidth=1)  # Draw vertical line

# Fit trend lines for specified periods
# Trend line for 1990-2011
years_1990_2011 = np.arange(1990, 2012)
contributions_1990_2011 = total_contributions.loc[1990:2011]
coeffs_1990_2011 = np.polyfit(years_1990_2011, contributions_1990_2011, 1)
poly_1990_2011 = np.poly1d(coeffs_1990_2011)
ax.plot(years_1990_2011, poly_1990_2011(years_1990_2011), 'b--', label='1990-2011 Trend')

# Trend line for 2011-2021
years_2011_2021 = np.arange(2011, 2022)
contributions_2011_2021 = total_contributions.loc[2011:2021]
coeffs_2011_2021 = np.polyfit(years_2011_2021, contributions_2011_2021, 1)
poly_2011_2021 = np.poly1d(coeffs_2011_2021)
ax.plot(years_2011_2021, poly_2011_2021(years_2011_2021), 'b--', label='2011-2021 Trend')

# Customize the plot
plt.title('Stacked Area Chart of Sector Contributions Over Time with Significant Inflection Points')
plt.xlabel('Year')
plt.ylabel('Contributions')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/michaelzimmermannstudhslu/BINA_2024_energy_research/main/data/bip_total.csv')
df = df.dropna()

# Initialize the linear regression model
model_population = LinearRegression()
model_gdp = LinearRegression()

# Prepare data for regression (reshape year data for fitting)
X = df['Jahr'].values.reshape(-1, 1)  # Features
y_population = df['Bevölkerung'].values  # Target for population
y_gdp = df['BIP'].values  # Target for GDP

# Fit the models
model_population.fit(X, y_population)
model_gdp.fit(X, y_gdp)

# Predict the values
y_population_pred = model_population.predict(X)
y_gdp_pred = model_gdp.predict(X)

fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot population data
color = 'tab:red'
ax1.set_xlabel('Jahr')
ax1.set_ylabel('Bevölkerung', color=color)
ax1.plot(df['Jahr'], df['Bevölkerung'], marker='o', color=color, label='Actual Bevölkerung')
ax1.plot(df['Jahr'], y_population_pred, color=color, linestyle='--', label='Regression Line - Bevölkerung')
ax1.tick_params(axis='y', labelcolor=color)

# Second y-axis for GDP
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('BIP', color=color)
ax2.plot(df['Jahr'], df['BIP'], marker='o', color=color, label='Actual BIP')
ax2.plot(df['Jahr'], y_gdp_pred, color=color, linestyle='--', label='Regression Line - BIP')
ax2.tick_params(axis='y', labelcolor=color)

# Title and legend
plt.title('Bevölkerung und BIP der Schweiz mit Regression (1980 - 2022)')
fig.tight_layout()

# Handling legends for both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left')
ax2.legend(lines + lines2, labels + labels2, loc='upper left')
plt.show()
```
